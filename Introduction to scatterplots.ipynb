# Introduction to scatterplots


the scatter plot is graph of the ordared pairs (x,y)
of numbers conisiting of the independent variabel x and the dependant y . 


```python
import matplotlib.pyplot as plt

data = {
    "student": ["A", "B", "C", "D", "E", "F"],
    "hour_of_study_x": [6, 2, 1, 5, 2, 3],
    "grade_y_percent": [82, 63, 57, 88, 68, 75]
}

```

```python
 
plt.scatter(x=data["hour_of_study_x"], y=data["grade_y_percent"])
plt.xlabel("Hours of Study")
plt.ylabel("Grade (%)")
plt.title("Scatter Plot of Study Hours vs Grade")
plt.show()
```

```python
import numpy as np

# Generate sample data for a linear relationship
np.random.seed(42)  # Set seed for reproducibility

# Create a range of x values
x_values = np.linspace(0, 10, 50)

# Generate y values as a linear function of x with some noise
# For example, y = 2x + 1 + noise
y_values = 2 * x_values + 1 + np.random.normal(scale=2, size=len(x_values))

# Plot the scatter plot
import matplotlib.pyplot as plt

plt.scatter(x_values, y_values)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot of Linear Relationship")
plt.show()
```

```python
import numpy as np
np.random.seed(42)  
x_values = np.linspace(0, 10, 50)
y_values = -2 * x_values + 1 + np.random.normal(scale=2, size=len(x_values))
import matplotlib.pyplot as plt
plt.scatter(x_values, y_values)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot of Negative Linear Relationship")
plt.show()
```

```python
import numpy as np
np.random.seed(42)   
x_values = np.linspace(-5, 5, 50)
y_values = x_values**2 + 2 * x_values + 1 + np.random.normal(scale=5, size=len(x_values))

import matplotlib.pyplot as plt

plt.scatter(x_values, y_values)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot of Curvilinear Relationship (polynomial)")
plt.show()
```
